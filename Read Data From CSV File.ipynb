```python
#Import Library
import pandas as pd
```

```python
#Create Data Frame and store the CSV File data into it
#Scenario 1) We have columns in the CSV File
df = pd.read_csv("Test.csv")
df
```

```python
#Scenario 2) We do not have columns in CSV File and want to give explicit list of Columns
df = pd.read_csv("Test.csv", names=['ID', 'Emp_Name', 'Salary'])
df
```

```python
#Scenario 2) We want to read only few rows from CSV File
df = pd.read_csv("Test.csv",nrows=5)
df
```

```python
#Scenario 4) We want to read only selected list of columns
df = pd.read_csv("Test.csv", usecols=['ID', 'Name'])
df
```
